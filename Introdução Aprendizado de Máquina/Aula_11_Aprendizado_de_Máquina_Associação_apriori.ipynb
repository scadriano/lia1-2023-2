```python
# Biblioteca para trabalharmos o .csv
import pandas as pd

# Dessa biblioteca importamos essa função
from apyori import apriori
```

## Base de dados mercado 1

```python
mercado_df = pd.read_csv('/content/mercado.csv', header = None)
# Observar as transações - NaN -> valor nulo
mercado_df
```

```python
mercado_df.shape
```

```python
from pandas.core.reshape.merge import merge_asof
# Colocar no formato de lista
transacoes = []
for i in range(len(mercado_df)):
  #print(i)
  #print(base_mercado1.values[i, 0])
  transacoes.append([str(mercado_df.values[i, j]) for j in range(mercado_df.shape[1])])
```

```python
transacoes
```

```python
# Lista no Python é uma necessidade da biblioteca apyori
type(transacoes)
```

```python
# Variável para receber as regras - passar parâmetros
regras = apriori(transacoes, min_support = 0.3, min_confidence = 0.8, min_lift = 2)

# Transformar a regra em formato de lista
resultados = list(regras)
```

## Base de dados mercado 2

```python
mercado2_df = pd.read_csv('mercado2.csv', header=None)
mercado2_df
```

```python
mercado2_df.shape
```

```python
transacoes2 = []
for i in range(mercado2_df.shape[0]):
  transacoes2.append([str(mercado2_df.values[i, j]) for j in range(mercado2_df.shape[1])])
```

```python
transacoes2
```

```python
# Parâmetros iniciais -> 0
#regras2 = apriori(transacoes2, min_support = 0.015, min_confidence = 0.77, min_lift = 8)

# Parâmetros mais baixos para termos algumas regras -> 160
regras2 = apriori(transacoes2, min_support = 0.003, min_confidence = 0.2, min_lift = 3)

resultados2 = list(regras2)

len(resultados2)
```

```python
# Retornar os resultados em Dataframe
A = []
B = []
suporte = []
confianca = []
lift = []

for resultado in resultados:
  s = resultado[1]
  result_rules = resultado[2]
  for result_rule in result_rules:
    a = list(result_rule[0])
    b = list(result_rule[1])
    c = result_rule[2]
    l = result_rule[3]
    A.append(a)
    B.append(b)
    suporte.append(s)
    confianca.append(c)
    lift.append(l)
```

```python
rules_df = pd.DataFrame({'A': A, 'B': B, 'suporte': suporte, 'confianca': confianca, 'lift': lift})

# Cenário 1
#rules_df.sort_values(by = 'suporte', ascending = False)

# Cenário 2
#rules_df.sort_values(by = 'confianca', ascending = False)

# Cenário 3
rules_df.sort_values(by = 'lift', ascending = False)
```
